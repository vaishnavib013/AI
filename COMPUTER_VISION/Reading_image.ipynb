# By ratio

```python
import cv2

#Asumme img_rgb is already loaded with an image
w_ratio=0.5#sacling factor for width
h_ratio=0.5#sacling factor for height

#Resize the image
new_img=cv2.resize(img_rgb,(0,0),fx=w_ratio,fy=h_ratio)
'''
cv2.resize:
This is afunction from openCV lib that resizes to a specific size.

img_rgb:
This is the input image that you want to resize.
Tt is assumed to be in RGB format.

(0,0):
This is the desired size of the output image.
Setting bith dimensions to 0 means that the size will be determined by the scaling factor provided
'''
plt.imshow(new_img)
```

# Flipping image

```python
%matplotlib inline
```

```python
#Along central x-axis
new_img=cv2.flip(new_img,0)
plt.imshow(new_img)
'''
0:
This is t he flip code that specifies how thw image should be flipped:
0 means  flipping around the x-axis(vertical flip).
1 means  flipping around the y-axis(horizontal flip).
-1 means  flipping around both the axis(both vertical and horizontal flip).
'''
```

```python
#Along central y-axis
new_img=cv2.flip(new_img,1)
plt.imshow(new_img)
```

```python
#Along both axis
new_img=cv2.flip(new_img,-1)
plt.imshow(new_img)
```

```python
img_bgr=cv2.imread('C:/6-computer_vision/00-puppy.jpg')
plt.imshow(img_bgr)
```

the image has been correctly loaded by opencvas a numpy array,but the color of each pixel has been sorted as 
BGR.Matplotlib's plot expect an RGB image so,for a correct display of an image it is necessary to swap those
channels.this operation can be done either by using openCV conversion functions cv2.cvtColor() or by working
directly with the  numpy array.

```python
img_rbg=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rbg)
```

```python
img_grey=cv2.imread('C:/6-computer_vision/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_grey)
```

```python

import cv2
#assume img_rgb is alreasy loaded with an image
w_ratio =0.5#scaling factor for width
h_ratio=0.5

#resizing the image
new_img=cv2.resize(img_rbg,(0,0),fx=w_ratio,fy=h_ratio)
"""this is a function from the openCV library that resizes
an image to a specified size
"""
```

```python
plt.imshow(new_img)
```

```python
%matplotlib inline
```

```python
#along central x axis
new_img=cv2.flip(new_img,0)
plt.imshow(new_img)
```

```python
"""0:
this is the flip code thst specofies hoqw the image should be flipped:
0 means flipping around the x-axis
1 means flipping around y axis
-1 means flipping aorund both the axes (both vertical and horizontal flip)"""
```

```python
#along central y-axis
new_img=cv2.flip(new_img,1)
plt.imshow(new_img)#here we will be getting mirror image across y axis
```

```python
#flipping across both the axis ie x and y axis
new_img=cv2.flip(new_img,-1)
plt.imshow(new_img)
```

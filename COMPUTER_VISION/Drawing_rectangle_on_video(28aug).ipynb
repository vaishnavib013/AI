```python
import cv2

cap=cv2.VideoCapture(0)

#automatically grab width and height from the video feed
#(return float which we need to convert into integer for later on!)
width =int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height =int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

#we're using //here because in pyhton //allows for int classical division,
#because we can't pass a float to the cv2.rectangle function

#coordinates for rectangle ,centre point rectangle
x=width//2
y=height//2

#width and height adding 1/4 of width and height from the centre
w=width//4
h=height//4

while True:
    #capture frame-by-frame
    ret,frame=cap.read()
    
    #draw a rectangle on stream
    cv2.rectangle(frame,(x,y),(x+w,y+h),color=(0,0,255),thickness=4)
    
    #Display the resulting frame
    cv2.imshow('frame',frame)
    
    #this is the command let's us quit with the "q" button on a keyboard
    #simply pressing X on the window wont work!
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
#when everthing is done ,release the capture
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
#ceate  a function based on a CV2 Event (left button click)
#havent drawn anything yet
#global variables
pt1=(0,0)
pt2=(0,0)
topLeft_clicked=False
botRight_clicked=False

#mouse callback function,tjis is resetting the rectangle
'''This function is trigged whenever a mouse event occur in the specified window .It checks if the left mouse button is clicked 
(cv2.EVENT_LBUTTONDOWN)
If both corners have ALREADY been clicked ,it resets the rectangle.
If the top-left corner hasn't been clicked yet,
it set pt1 to the current mouse position and sets topleft
it set pt2 to the current mouse position and sets botright_clicked to TTrue.
'''
def draw_rectangle(event,x,y,flags,param):
    global pt1,pt2,topLeft_clicked,botRight_clicked
    
    #get mouse click
    if event==cv2.EVENT_LBUTTONDOWN:
#when you draw a fresh new rectangle ,it resets older coordinates
        if topLeft_clicked == True and botRight_clicked == True:
            topLeft_clicked = False
            botRight_clicked = False
            pt1 = (0,0)
            pt2 = (0,0)

        if topLeft_clicked == False:
            pt1 = (x,y)
            topLeft_clicked = True

        elif botRight_clicked == False:
            pt2 = (x,y)
            botRight_clicked = True

            '''
            Checking if topLeft clicked is False:
            '''

cap = cv2.VideoCapture(0)

    #create a named window for connection
cv2.namedWindow('Test')

    #Bind draw_rectangle function to mouse clicks
cv2.setMouseCallback('Test',draw_rectangle)
""" 
"""
while True:
        #capture frame-by-frame
    ret,frame=cap.read()

        #draw a rectangle on stream
    if topLeft_clicked:
        cv2.circle(frame,center=pt1,radius=5,color=(0,0,255),thickness=-1)

        #drawing the rectangle
    if topLeft_clicked and botRight_clicked:
        cv2.rectangle(frame,pt1,pt2,(0,0,255),2)

        #Display the resulting frame
    cv2.imshow('Test',frame)

        #this is the command let's us quit with the "q" button on a keyboard
        #simply pressing X on the window wont work!
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    #when everthing is done ,release the capture
cap.release()
cv2.destroyAllWindows()
```
